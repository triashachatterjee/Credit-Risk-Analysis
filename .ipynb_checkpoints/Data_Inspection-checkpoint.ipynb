```python
# Defining the target variable (0: Good/Non-Default, 1: Bad/High Risk)
# We treat 'Charged Off', 'Default', and serious delinquencies as 'Bad'.
# We exclude 'Current' and 'In Grace Period' loans for classification, as their final status is unknown,
# but for a comprehensive risk model, they might be treated as 'Good' if we are predicting "Default at Issue Time".
# A conservative approach for LTV (Loan-To-Value) prediction often focuses on settled loans.
```

```python
bad_loan_statuses = ['Charged Off', 'Default', 'Late (31-120 days)',
                     'Late (16-30 days)', 'Does not meet the credit policy. Status:Charged Off']
good_loan_statuses = ['Fully Paid', 'Does not meet the credit policy. Status:Fully Paid']
unknown_statuses = ['Current', 'In Grace Period', 'Issued', 'Not Yet Issued']
```

```python
# Filter to only include loans that have been settled (Fully Paid or Charged Off)
# This is common practice to avoid data leakage from "Current" loans.
df_target = df[df['loan_status'].isin(bad_loan_statuses + good_loan_statuses)].copy()
```

```python
df_target['is_default'] = np.where(df_target['loan_status'].isin(bad_loan_statuses), 1, 0)
```

```python
print(f"\nTotal number of settled loans used for modeling: {df_target.shape[0]}")
print(f"Default Rate (1) in the filtered dataset: {df_target['is_default'].mean():.2%}")
```

```python
#21.48%  of the total loan are defaulters
```

```python
# Convert issue_d to datetime and calculate loan age in months (relative to today/last date in dataset)
df_target['issue_d'] = pd.to_datetime(df_target['issue_d'], format='%b-%Y')
```

```python
# Assuming today is the latest date in the dataset for age calculation
latest_date = df_target['issue_d'].max()
df_target['loan_age_months'] = (latest_date.year - df_target['issue_d'].dt.year) * 12 + \
                               (latest_date.month - df_target['issue_d'].dt.month)
```

```python
columns_to_drop = [
    # Identifiers & Text (Irrelevant/High Cardinality)
    'id', 'member_id', 'url', 'desc', 'title', 'emp_title', 'zip_code',

    # Post-Loan Outcome / Data Leakage (MUST DROP)
    # These contain information only known AFTER the loan has been serviced or defaulted.
    'out_prncp', 'out_prncp_inv', 'total_pymnt', 'total_pymnt_inv',
    'total_rec_prncp', 'total_rec_int', 'total_rec_late_fee', 'recoveries',
    'collection_recovery_fee', 'last_pymnt_d', 'last_pymnt_amnt',
    'next_pymnt_d', 'last_credit_pull_d', 'last_fico_range_high',
    'last_fico_range_low', 'mths_since_last_major_derog',
    'total_rec_prncp', 'total_rec_int', 'total_rec_late_fee',

    # Hardship & Settlement (Post-issue events)
    'hardship_flag', 'hardship_type', 'hardship_reason', 'hardship_status',
    'deferral_term', 'hardship_amount', 'hardship_start_date', 'hardship_end_date',
    'payment_plan_start_date', 'hardship_length', 'hardship_dpd', 'hardship_loan_status',
    'orig_projected_additional_accrued_interest', 'hardship_payoff_balance_amount',
    'hardship_last_payment_amount', 'debt_settlement_flag', 'debt_settlement_flag_date',
    'settlement_status', 'settlement_date', 'settlement_amount', 'settlement_percentage',
    'settlement_term',

    # Policy and application details (Usually single values or redundant)
    'policy_code', 'application_type', 'pymnt_plan', 'funded_amnt', 'funded_amnt_inv', # funded_amnt is highly correlated with loan_amnt
]
```

```python
df_cleaned = df_target.drop(columns=columns_to_drop, errors='ignore')
```

```python
#Counting the missing values
```

```python
missing_counts = df_cleaned.isnull().sum()
missing_percent = (missing_counts / df_cleaned.shape[0])*100
```

```python
# Filter to show only columns with missing data
missing_df = pd.DataFrame({'Missing Count': missing_counts, 'Missing %': missing_percent})
missing_df = missing_df[missing_df['Missing Count'] > 0].sort_values(by='Missing %', ascending=False)
```

```python
print("\nTop 20 columns with missing values:")
print(missing_df.head(20))
```

```python
# Strategy Implementation Example: Drop columns with > 70% missing data
# (Most of these are secondary applicant or long-term history fields)
high_null_cols = missing_df[missing_df['Missing %'] > 70].index.tolist()
df_cleaned.drop(columns=high_null_cols, errors='ignore', inplace=True)
```

```python
print(f"\nDropped {len(high_null_cols)} columns due to high missing percentage (> 70%).")
print(f"Current number of columns: {df_cleaned.shape[1]}")
```
